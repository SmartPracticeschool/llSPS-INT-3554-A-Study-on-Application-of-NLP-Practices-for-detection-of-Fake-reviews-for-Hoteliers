```python
import os
os.chdir('/Users/obadaessa/Desktop/Internship/Data Sets')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data1 = pd.read_csv('Position_Salaries.csv')

data1

x = data1.iloc[:, 1:2].values # x = Level
y = data1.iloc[:, 2].values   # y = Salary
```

```python
x, y
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

linreg = LinearRegression()
linreg.fit(x, y)
```

```python
xtrain
```

```python
ytrain
```

```python
def vizlinear():
    plt.scatter(x, y, color='blue')
    plt.plot(x, linreg.predict(x), color='green')
    plt.title('Truth or bluff (Linear Regression)')
    plt.xlabel('Position Level')
    plt.ylabel('Salary')
    plt.show()
    return
vizlinear()
```

```python
# Fitting Poly Reg 

from sklearn.preprocessing import PolynomialFeatures
```

```python
polyreg = PolynomialFeatures(degree=4)
xpoly = polyreg.fit_transform(x)
polreg = LinearRegression()
polreg.fit(xpoly, y)
```

```python
def vizpoly():
    plt.scatter(x, y, color='blue')
    plt.plot(x, polreg.predict(xpoly), color='green')
    plt.title('Truth or bluff (Linear Regression)')
    plt.xlabel('Position Level')
    plt.ylabel('Salary')
    plt.show()
    return
vizpoly()
```

```python
def smoothpoly():
    xgrid = np.arange(min(x), max(x), 0.1)
    xgrid = xgrid.reshape(len(xgrid), 1)
    plt.scatter(x, y, color='blue')
    plt.plot(xgrid, polreg.predict(polyreg.fit_transform(xgrid)), color='green')
    plt.title('Truth or bluff (Linear Regression)')
    plt.xlabel('Position Level')
    plt.ylabel('Salary')
    plt.show()
    return
smoothpoly()
```
